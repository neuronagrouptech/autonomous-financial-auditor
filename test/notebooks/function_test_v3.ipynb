# S3 Boto Agent

```python
import boto3
import uuid
```

```python
session = boto3.Session(profile_name='finance-project')
client = session.client('bedrock-agent-runtime', region_name='us-east-1')
session_id = str(uuid.uuid4())
```

```python
response = client.invoke_agent(    agentId="XYKIHMJLKO",
    agentAliasId="FHELHMWAIJ",
    sessionId = session_id, 
    inputText='This is a test'
)
```

```python
response
```

```python
output = ""
for event in response["completion"]:
    if "chunk" in event:
        output += event["chunk"]["bytes"].decode("utf-8")
```
